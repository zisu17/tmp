```python
from urllib.request import urlopen
import bs4
import pandas as pd
```

```python
title=[]
for i in range(1,15):
    try:
        url="https://terms.naver.com/list.naver?cid=42883&categoryId=44358&page=" + str(i)
        dog_Info = urlopen(url)
        dog_url = bs4.BeautifulSoup(dog_Info,'html.parser')
        info = dog_url.findAll('div',{'class':'info_area'})
        for i in range(0,15):
            name = info[i].find('strong',{'class':'title'}).select('a')[0].text.split('[')[0].replace(' ',"")
            title.append(name)
    except:
        continue
print(title)
```
